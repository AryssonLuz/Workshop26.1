# **1) Numpy**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

temperatura = np.random.randint(20,35,size=7)
print(f'temperaturas: {temperatura}')

dias = np.array(['segunda','terça','quarta','quinta','sexta','sabado','domingo'])
print(f'dias: {dias}')
```

# **2) Pandas**

```python
dias_temperatura = pd.Series(temperatura,dias)
dias_temperatura

media = dias_temperatura.mean()
print(f'media: {media:.2f}')

maxima = dias_temperatura.max()
minima = dias_temperatura.min()

print(f'maxima: {maxima}')
print(f'minimo: {minima}')
```

# **3) Matplotlib.1**

```python
plt.plot(dias_temperatura)
plt.xlabel('dias')
plt.ylabel('temperatura')
plt.show()
```

# **3) Matplotlib.2**

```python
plt.bar(dias,temperatura)
plt.xlabel('dias')
plt.ylabel('temperatura')
plt.show()
```

# **3) Matplotlib.3**

```python
plt.scatter(dias,temperatura)
plt.xlabel('dias')
plt.ylabel('temperatura')
plt.show()
```
